<h3>–†—ñ—à–µ–Ω–Ω—è:</h3>

````python
def prime_order():
    try:
        k = int(input())
        if k <= 0:
            print("–ü–æ–º–∏–ª–∫–∞. –ó–Ω–∞—á–µ–Ω–Ω—è –ø–æ–≤–∏–Ω–Ω–æ –±—É—Ç–∏ > 0")
            return
        
        primes = []
        num = 2
        
        while len(primes) < k:
            is_prime = True
            for p in primes:
                if p * p > num:
                    break
                if num % p == 0:
                    is_prime = False
                    break
            if is_prime:
                primes.append(num)
            num += 1
        
        print(f"{k}-—Ç–µ –ø—Ä–æ—Å—Ç–µ —á–∏—Å–ª–æ - —Ü–µ {primes[-1]}")
        
    except ValueError:
        print("–ü–æ–º–∏–ª–∫–∞. –ó–Ω–∞—á–µ–Ω–Ω—è –ø–æ–≤–∏–Ω–Ω–æ –±—É—Ç–∏ —á–∏—Å–ª–æ–≤–µ")
        return
````


<h3>–¢–µ—Å—Ç—É–≤–∞–Ω–Ω—è:</h3>
