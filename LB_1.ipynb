Для створеної функції реалізувати 2-3 тестові приклади в комірках ноутбуку. 
Для всіх варіантів організувати контроль значень, що вводяться.
<h3>Умова алгоритму:</h3>
Вивести k-те за порядком просте число. Число k вводиться під час виконання скрипта.
Приклад:


```python
In [2]: task_5()
5
Out[2]: 5-те просте число - це 11

In [3]: task_5()
34
Out[3]: 34-те просте число - це 139
```


<h3>Рішення:</h3>

````python
def prime_order():
    try:
        k = int(input())
        if k <= 0:
            print("Помилка. Значення повинно бути > 0")
            return
        
        primes = []
        num = 2
        
        while len(primes) < k:
            is_prime = True
            for p in primes:
                if p * p > num:
                    break
                if num % p == 0:
                    is_prime = False
                    break
            if is_prime:
                primes.append(num)
            num += 1
        
        print(f"{k}-те просте число - це {primes[-1]}")
        
    except ValueError:
        print("Помилка. Значення повинно бути числове")
        return
````
